```python
import re
import pandas as pd
import unicodedata

def clean_resume(text):
    if not isinstance(text, str):
        return ""
    
    # Normalize Unicode characters (NaÃ¯ve → Naive)
    text = unicodedata.normalize("NFKD", text).encode("ascii", "ignore").decode("utf-8")
    
    # Lowercase all text (optional)
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    
    # Remove mentions and hashtags (general purpose)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#\w+', '', text)
    
    # Remove digits
    text = re.sub(r'\d+', '', text)
    
    # Remove punctuations and special characters
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove weird bullets or artifacts
    text = re.sub(r'[•\u2022]', '', text)
    
    # Normalize all types of line breaks (\r, \n, \r\n, multiple)
    text = re.sub(r'(\r\n|\r|\n)+', ' ', text)
    
    # Remove single letters
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)
    
    # Replace multiple spaces with single space
    text = re.sub(r'\s+', ' ', text)
    
    # Trim leading/trailing spaces
    text = text.strip()
    
    return text

```

```python
resume_info['Resume'].apply(lambda x: clean_resume(x))
resume_info['Resume'][0]
```
